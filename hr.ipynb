```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train_processed, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train_resampled, y_train_resampled)

y_pred_knn = knn_model.predict(X_test_processed)

print("K-Nearest Neighbors Model Evaluation (with Oversampled Training Data):")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

rf_oversampled_model = RandomForestClassifier(
    n_estimators=50,
    max_depth=8,
    random_state=42
)

rf_oversampled_model.fit(X_train_resampled, y_train_resampled)

y_pred_rf_oversampled = rf_oversampled_model.predict(X_test_processed)

print("Random Forest Classifier Evaluation (with Oversampled Training Data):")
print("Accuracy:", accuracy_score(y_test, y_pred_rf_oversampled))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_oversampled))
print("Classification Report:\n", classification_report(y_test, y_pred_rf_oversampled))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

log_reg_oversampled = LogisticRegression(max_iter=1000, random_state=42)

log_reg_oversampled.fit(X_train_resampled, y_train_resampled)

y_pred_log_reg_oversampled = log_reg_oversampled.predict(X_test_processed)

print("Logistic Regression Evaluation (with Oversampled Training Data):")
print("Accuracy:", accuracy_score(y_test, y_pred_log_reg_oversampled))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_log_reg_oversampled))
print("Classification Report:\n", classification_report(y_test, y_pred_log_reg_oversampled))
```

As u can see the best model is Knearestneighbour.
